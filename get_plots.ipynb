```python
# Cluster and Plot heatmap from a matrix

from __future__ import division
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import squareform
from scipy.cluster.hierarchy import dendrogram,linkage
import pylab

import matplotlib as mpl
# mpl.rc('image', cmap='jet')

# 1) ####
# Read csv file into pandas dataframe
# df = pd.read_csv('~/GT/gta_revise13/analysis/machine_learning/dataset/mechanism_set/ccmpred/p.mat',sep='\t',header=None)
df1 = pd.read_csv('~/GT/gta_revise13/analysis/tree_comparison/pdb_clusters/rev13_cealign.txt.e1',sep=',')
df=df1.iloc[:,1:]

# Convert rmsd distance matrix to a 1d distance
# test = [[0,1,2],[1,0,1],[2,1,0]]
# print(test)
dist=squareform(df)
# print(dist)

# Generate clusters
clust=linkage(dist,'ward')
# print(clust)
# fig=plt.figure(figsize=(25,10))
cols=list(df.columns.values)

# Plot dendrogram
fig = plt.figure(figsize=(35,25))
axdendro = fig.add_axes([0.01,0.1,0.2,0.8])
dn=dendrogram(clust, orientation='left', color_threshold=7, above_threshold_color='black', leaf_font_size=22,labels=cols)
# print(dn['ivl'])

axdendro.set_xticks([])
# axdendro.set_yticks([])
# axdendro.set_xticklabels(cols)

# Reorder data matrix based on the dendrogram indices
index = dn['leaves']
cols=list(df.columns.values)
cols=[cols[i] for i in index]
df=df.reindex(index)
df=df.reindex(columns=cols)

# Plot heatmap of the data matrix
axmatrix = fig.add_axes([0.31,0.1,0.6,0.8])
# im = sns.heatmap(df, linewidth=0.5)
im = axmatrix.matshow(df, aspect='auto', origin='lower',cmap="YlGnBu")
axmatrix.set_xticks([])
axmatrix.set_yticks([])

# Plot colorbar.
axcolor = fig.add_axes([0.92,0.1,0.02,0.8])
plt.colorbar(im, cax=axcolor) #.set_label(weight='bold')
# cb.set_fontsize('20')
# Display and save figure.
fig.show()
fig.savefig("pdbDendrogram_rev13_cealign_ward.png")

# fig = plt.figure(figsize=(12, 20))
# ax1 = fig.add_subplot(211)
# ax2 = fig.add_subplot(212)
# dendrogram(clust, ax=ax1)
# ax2.pcolor(df, cmap=plt.cm.Reds)

# clust=sns.clustermap(df,metric="euclidean",standard_scale=1,method="ward")
# from sklearn.cluster import AgglomerativeClustering
# cluster = AgglomerativeClustering(n_clusters=20, affinity='euclidean', linkage='ward')  
# cluster.fit_predict(df) 
```

```python
import scipy
import pylab
import scipy.cluster.hierarchy as sch

# Generate features and distance matrix.
x = scipy.rand(40)
D = scipy.zeros([40,40])
for i in range(40):
    for j in range(40):
        D[i,j] = abs(x[i] - x[j])

# Compute and plot dendrogram.
fig = pylab.figure()
axdendro = fig.add_axes([0.09,0.1,0.2,0.8])
Y = sch.linkage(D, method='centroid')
Z = sch.dendrogram(Y, orientation='right')
axdendro.set_xticks([])
axdendro.set_yticks([])

# Plot distance matrix.
axmatrix = fig.add_axes([0.3,0.1,0.6,0.8])
index = Z['leaves']
D = D[index,:]
D = D[:,index]
im = axmatrix.matshow(D, aspect='auto', origin='lower')
axmatrix.set_xticks([])
axmatrix.set_yticks([])

# Plot colorbar.
axcolor = fig.add_axes([0.91,0.1,0.02,0.8])
pylab.colorbar(im, cax=axcolor)

# Display and save figure.
fig.show()

fig.savefig("pdb_dendrogram_ward.png")
```
